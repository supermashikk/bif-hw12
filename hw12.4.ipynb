```python
import pandas as pd
from scipy.stats import ttest_ind, ttest_rel #непарный и парный тесты соответственно


df = pd.read_csv('https://raw.githubusercontent.com/s-a-nersisyan/HSE_bioinformatics_2021/master/seminar12/colon_cancer_tumor_vs_normal_paired_FPKM.tsv', sep = '\t', index_col = 0)

#непарный тест
df['p-value-unpair'] = [ttest_ind(df.loc[gene].iloc[0:5], df.loc[gene].iloc[5:10])[1] for gene in df.index]  

#парный тест
df['p-value-pair'] = [ttest_rel(df.loc[gene].iloc[0:5], df.loc[gene].iloc[5:10])[1] for gene in df.index]
```

сначала идут опухоли, потом - здоровые ткани

```python
df
```

```python
df_pair = df['p-value-pair']
df_unpair = df['p-value-unpair']

df_pair = df_pair[df_pair < 0.05].sort_values()
df_unpair = df_unpair[df_unpair < 0.05].sort_values()
```
